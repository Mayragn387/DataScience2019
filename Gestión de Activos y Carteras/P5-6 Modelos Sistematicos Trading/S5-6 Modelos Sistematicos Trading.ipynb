Complementando la practica de las sesiones anteriores, se aplicaron estrategias de predicción de rendimiento a los activos como:
- Trend Following, es una estrategia comercial según la cual uno debe comprar un activo cuando su tendencia de precios aumenta y vender cuando su tendencia baja, esperando que continúen los movimientos de precios. Se basa en dos supuestos: asume que el comportamiento histórico del precio de un activo es el mejor indicador para predecir el comportamiento futuro,y los activos que mejor comportamiento han tenido en el pasado van a continuar teniéndolo en el futuro y viceversa. Esto implica una estrategia de comprar el activo cuanto más caro este y venderlo cuando más barato esté. Es lo que popularmente se denomina “momentum”.


- Mean Reversion, es totalmente diferente a las demás: Intentan comprar el activo cuanto más barato este y venderlo cuanto más caro. El operador que utilizar la estrategia de reversión a la media apuesta a que un activo que está subiendo de precios va a empezar a bajar, o que un activo con comportamiento bajista va a empezar a subir. La clave entonces es estar siempre atentos y en busca de potenciales oportunidades de reversión.

- Carry, se basa en una situación en la que un trader pide prestado, o vende, una divisa con un tipo de interés bajo para comprar otra divisa con un tipo de interés más alto. Llamamos “carry” al retorno o beneficio / coste que produce un activo mientras lo tenemos en posesión. Un carry trade positivo consiste en pedir prestado una divisa con un tipo de interés bajo mientras se compra otra divisa con un tipo de interés más alto. El trader recibirá entonces pagos de intereses basados en la diferencia de tipos de interés entre las dos divisas y el tamaño de su posición.

- Evaluación con relación a un indicador Macro, se evalua el rendimiento de un activo con respecto a un indicador macro, para ello se considera la correlación, si sigue el mismo comportamiento del mercado se emite la señal 1, en caso contrario la señal -1. Hemos utilizado el indicador __Indice Sorpresas Citi__ .

## Modelos de Prediccion de Rendimientos

* Trend Following: El proceso realizado es calcular primero las medias móviles a 10, 30, 50 y 100 días, ésto permitirá estimar si la tendencia del activo será en alza o baja a Corto, Mediana y Largo Plazo. Se crearan tres señales: Corto Plazo, será 1 (Buy) en caso de que la media móvil de 10 días sea mayor a 30 días y en caso contrario sera -1(Sell), se reliza lo mismo a Mediano Plazo, se comparan la media móvil de 30 días con respecto a la SMA de 50 días, y Largo Plazo, se realiza el mismo análisis con respecto a las medias móviles de 50 y 100 días. Aqui aplicamos pesos equitativos (33.3%). También se aplicó la estrategia de Breakouts, donde se estimaron los valores maximos y minimos en ventanas de 100 días (los llamados Resistencia y Soporte) si el indice es mayor al valor máximo(Resistencia) se emite una señal 1(Buy), en cambio se emite señal -1 (Sell) si es menor al valor minimo (Soporte), en otros casos es 0. Ambas señales se ponderan 50% cada una y se genera una señal final. Se genera un indice final con el indice del activo*(1+Señal*Porcentaje_Rentabilidad).

* Mean Reversion: Se calculan las diferencias diarias (Movimientos) y se clasifican en Ganancias y Perdidas. Se genera la media movil de estos valores a 14 días y se divide la SMA de las ganancias/la SMA de las perdidas (RS). Este ratio sera utilizado para calcular un nuevo indicador calculado de la siguiente manera 100-(100/(1+RS)). Si es menor a 30 se emite la señal 1 (Buy) y si es mayor a 70 se emite a -1 (Sell), en otros casos es 0. Se genera finalmente el indice final considerando la señal de la siguiente manera: indice del activo*(1+Señal*Porcentaje_Rentabilidad).

* Relación con la indicador Macro: Se calcula los coeficientes de correlacion entre el indice del activo y el indicador macro en ventanas moviles de 252 días (anualizados). Se compara el indicador macro y este coeficiente de correlacion, si ambos son positivos o ambos negativos, se emite la señal 1(Buy), en caso contrario, si son diferentes se emite la señal -1(Sell). En otros casos se emite la señal 0.Se genera finalmente el indice final considerando la señal de la siguiente manera: indice del activo*(1+Señal*Porcentaje_Rentabilidad).

* Carry Divisas: en esta estrategia se compara los valores a tipo anualizados, si la primera divisa (AUD-USD) es mayor a la segunda divisa (JPY-USD) la señal emitida es 1, en caso contrario es -1. Si la señal es 1, se calcula el indice del carry de la siguiente manera indice del activo * (1+ (Tipo_Anualizado_Primera_Divisa-Tipo_Anualizado_Segunda_Divisa)/100*(1/365)), en el caso de ser -1, la formula es indice del activo * (1+ (Tipo_Anualizado_Segunda_Divisa-Tipo_Anualizado_Primera_Divisa)/100*(1/365)). Se calcula la variacion del nuevo indice. Por otro lado se genera los nuevos indices segun la señal de cada divisa: si es señal 1, los indices se incrementan por la tasa de (1+Rentabilidad_Primera_Divisa) y la tasa de la segunda divisa es(1-Rentabilidad_Segunda_Divisa). Si es señal -1, los indices se incrementan por la tasa de (1-Rentabilidad_Primera_Divisa) y la tasa de la segunda divisa es(1+Rentabilidad_Segunda_Divisa) y se calcula la tasa de variación. La rentabilidad es la suma de las tres tasas de variación.

#### Reversión con Precios (Mean Reversion)

```python
prices_df['DJI_mov'] = prices_df['DJI_IDX'].diff()
gains = []
gains.append(np.nan)
loss = []
loss.append(np.nan)
for i in range(1,prices_df.shape[0]):
    if prices_df['DJI_mov'].iloc[i]>0:
        gains.append(prices_df['DJI_mov'].iloc[i])
        loss.append(np.nan)
    else:
        if prices_df['DJI_mov'].iloc[i]<0:
            gains.append(np.nan)
            loss.append(-1*prices_df['DJI_mov'].iloc[i])
        else:
            gains.append(np.nan)
            loss.append(np.nan)
                       
prices_df['DJI_gain'] = gains
prices_df['DJI_loss'] = loss
```

```python
def risk_contribution(w,cov):
    """
    Compute the contributions to risk of the constituents of a portfolio, given a set of portfolio weights and a covariance matrix
    """
    total_portfolio_var = erk.portfolio_vol(w,cov)**2
    # Marginal contribution of each constituent
    marginal_contrib = cov@w
    risk_contrib = np.multiply(marginal_contrib,w.T)/total_portfolio_var
    return risk_contrib

def to_percent(x):
    try: f_string = '{:.2%}'.format(x)
    except: f_string = x
    return f_string

def to_decimal(x):
    try:
        if x >= 100:
            f_string = '{0:,.0f}'.format(x)
        elif x >= 10:
            f_string = '{0:,.1f}'.format(x)
        else:
            f_string = '{0:,.2f}'.format(x)
            
    except: f_string = x
    return f_string

def annual_volatility(df_single_returns):
    """
    Determines the annual volatility of a strategy.

    Parameters
    ----------
    df_single_returns : pd.Series or np.ndarray
        Periodic returns of the strategy, noncumulative.

    Returns
    -------
    float, np.ndarray
        Annual volatility.
    """

    if len(df_single_returns) < 2:
        return np.nan

    std = df_single_returns.std(ddof=1)

    volatility = std * (252 ** (1.0 / 2))

    return volatility.astype(np.float)
def var_pct(df_single_returns):
    """
    Determines the annual var of a strategy.

 

    Parameters
    ----------
    df_single_returns : pd.Series or np.ndarray
        Periodic returns of the strategy, noncumulative.

 

    Returns
    -------
    float, np.ndarray
        Annual var (%).
    """

 

    if len(df_single_returns) < 2:
        return np.nan

 

    std = df_single_returns.std(ddof=1)
    vol = std * (252 ** (1.0 / 2))
    t=252
    ci=0.95


    var =  vol*scs.norm.cdf(scs.norm.ppf(1-ci))/np.sqrt(t) #scs.norm.ppf(1-ci, mean, vol) * np.sqrt(t)
    return var.astype(np.float)

def max_drawdown(df_returns):
    max_dd = (np.cumprod(1+df_returns)/np.maximum.accumulate(np.cumprod(1+df_returns))-1).min()
    return max_dd

def cal_max_dd(df_single_return):
    """
    Determines the maximum drawdown of a strategy.

    Parameters
    ----------
    df_single_return :
        Daily returns of the strategy, noncumulative.

    Returns
    ----------
    float
        Maximum drawdown.
    """
    if len(df_single_return) < 1:
        return np.nan

    df_perform_equity_curve = (1. + df_single_return).cumprod()
    df_perform_cum_max = df_perform_equity_curve.cummax()
    # drawdown series
    df_perform_drawdown = df_perform_equity_curve / df_perform_cum_max - 1
    max_dd = df_perform_drawdown.min()
    return max_dd

def annual_return(df_single_return, period):
    """Determines the mean annual growth rate of returns.

    Parameters
    ----------
    df_single_return : pd.Series or np.ndarray
        Periodic returns of the strategy, noncumulative.
    period : str, optional
        Defines the periodicity of the 'returns' data for purposes of
        annualizing. Value ignored if `annualization` parameter is specified.
        Defaults are:
            'monthly':12
            'weekly': 52
            'daily': 252

    Returns
    -------
    float
    """

    if len(df_single_return) < 1:
        return np.nan

    num_years = float(len(df_single_return)) / period

    # Pass array to ensure index -1 looks up successfully.
    df_cum = (df_single_return + 1).cumprod(axis=0) - 1
    f_annual_return = (1. + df_cum.iloc[-1,:]) ** (1. / num_years) - 1

    return f_annual_return.apply(to_percent)

def to_summary(df_returns):
    # Daily return calcs
    df_summary = pd.DataFrame(columns=df_returns.columns)
    df_summary.loc['Summary Stats:'] = ''
    MOIC = np.prod(1+df_returns)
    CAGR = MOIC**(periodicity/len(df_returns))-1
    df_summary.loc['Annualized Return'] = annual_return(df_returns, 252)
    df_summary.loc['Sharpe'] = CAGR/(np.std(df_returns)*np.sqrt(periodicity))
    df_summary.loc['Volatility'] = annual_volatility(df_returns) 
    df_summary.loc['VaR'] = var_pct(df_returns).apply(to_percent)

    df_summary.loc['Sortino'] = CAGR/(np.std(np.clip(df_returns, a_max=0, a_min=None))*np.sqrt(periodicity))
    df_summary.loc['Max Drawdown'] = cal_max_dd(df_returns).apply(to_percent)

    df_summary = df_summary.applymap(to_decimal)
    
    # Monthly return calcs
    df_summary.loc['Monthly Perf. Metrics:'] = ''
    df_months = pd.DataFrame(columns=df_returns.columns)
    df_returnsM = (1+df_returns).resample('M').prod()-1
    MOIC_M = np.prod(1+df_returnsM)
    CAGR_M = MOIC_M**(12/len(df_returnsM))-1
    df_months.loc['Sharpe'] = CAGR_M/(np.std(df_returnsM)*np.sqrt(12))
    df_months.loc['Sortino'] = CAGR_M/(np.std(np.clip(df_returnsM, a_max=0, a_min=None))*np.sqrt(12))
    
    if len(df_returnsM) >= 36: df_months.loc['Calmar'] = -(np.prod(1+df_returnsM.iloc[-12*3:])**(1/3)-1)\
                                                         / max_drawdown(df_returnsM.iloc[-12*3:])
    
    df_months = df_months.applymap(to_decimal)
    df_summary = df_summary.append(df_months)
    
    # Annuals
    df_summary.loc['Annual Returns:'] = ''
    df_annuals = df_returns.groupby(df_returns.index.year).apply(lambda x: np.prod(1+x)-1).applymap(to_percent)
    df_summary = df_summary.append(df_annuals)

    # Drawdowns
    df_summary.loc['Annual Max DD:'] = ''    
    df_maxdd = df_returns.groupby(df_returns.index.year).apply(cal_max_dd).applymap(to_percent)
    df_summary = df_summary.append(df_maxdd)
    return df_summary
```

Tenemos 4 activos según las estrategias de rendimiento:

Para la construcción de la cartera nos basamos en obtener dos objetivos:
- Máximo indice de Sharpe.
- Mínima Volatilidad anualizada.
Se crearon las funciones `portfolio_annualised_performance` (para obtener la performance de la cartera) y `random_portfolios` para ejecutar la simulación de diferentes asignaciones (o pesos) de los activos.

```python
# "portfolio_annual_performance" function will calculate:
# - the returns
# - and volatility
# To make it as an annualised calculation, 252 is taken into account as the number of trading days in one year.

def portfolio_annualised_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns*weights ) *252
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    return std, returns
```

```python
# "random_portfolios" function will generate:
# - portfolios with random weights assigned to each stock
# Where:
#     - num_portfolios argument dictates how many random portfolios are generated.
    
def random_portfolios(num_portfolios, mean_returns, cov_matrix, risk_free_rate):
    results = np.zeros((3,num_portfolios))
    #results = np.zeros((returns.shape[1],num_portfolios))
    weights_record = []
    #for i in xrange(num_portfolios):
    for i in range(num_portfolios):
        weights = np.random.random(returns.shape[1])
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_std_dev, portfolio_return = portfolio_annualised_performance(weights, mean_returns, cov_matrix)
        results[0,i] = portfolio_std_dev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - risk_free_rate) / portfolio_std_dev
    return results, weights_record
```

La función `display_simulated_ef_with_random` realiza lo siguiente:
- Genera una cartera aleatoria y obtiene los resultados (rentabilidad de la cartera, volatilidad de la cartera, relación de Sharpe de la cartera) y ponderaciones para el resultado correspondiente.

- Todas las carteras generadas aleatoriamente también se trazarán con un mapa de color aplicado a ellas según la relación de Sharpe. Cuanto más azul, mayor relación de Sharpe.

- Y para estas dos carteras óptimas (relación máxima de nitidez y volatilidad mínima), también mostrará cómo asigna el presupuesto dentro de la cartera.

```python
def display_simulated_ef_with_random(mean_returns, cov_matrix, num_portfolios, risk_free_rate):
    results, weights = random_portfolios(num_portfolios,mean_returns, cov_matrix, risk_free_rate)
    
    max_sharpe_idx = np.argmax(results[2])
    sdp, rp = results[0,max_sharpe_idx], results[1,max_sharpe_idx]
    max_sharpe_allocation = pd.DataFrame(weights[max_sharpe_idx],index=df.columns,columns=['allocation'])
    max_sharpe_allocation.allocation = [round(i*100,2)for i in max_sharpe_allocation.allocation]
    max_sharpe_allocation = max_sharpe_allocation.T
    
    min_vol_idx = np.argmin(results[0])
    sdp_min, rp_min = results[0,min_vol_idx], results[1,min_vol_idx]
    min_vol_allocation = pd.DataFrame(weights[min_vol_idx],index=df.columns,columns=['allocation'])
    min_vol_allocation.allocation = [round(i*100,2)for i in min_vol_allocation.allocation]
    min_vol_allocation = min_vol_allocation.T
    
    print("-"*80)
    print("Maximum Sharpe Ratio Portfolio Allocation\n")
    print("Annualised Return:", round(rp,2))
    print("Annualised Volatility:", round(sdp,2))
    print("\n")
    print(max_sharpe_allocation)
    print("-"*80)
    print("Minimum Volatility Portfolio Allocation\n")
    print("Annualised Return:", round(rp_min,2))
    print("Annualised Volatility:", round(sdp_min,2))
    print("\n")
    print(min_vol_allocation)
    
    plt.figure(figsize=(10, 7))
    plt.scatter(results[0,:],results[1,:],c=results[2,:],cmap='YlGnBu', marker='o', s=10, alpha=0.3)
    plt.colorbar()
    plt.scatter(sdp,rp,marker='*',color='r',s=500, label='Maximum Sharpe ratio')
    plt.scatter(sdp_min,rp_min,marker='*',color='g',s=500, label='Minimum volatility')
    plt.title('Simulated Portfolio Optimization based on Efficient Frontier')
    plt.xlabel('annualised volatility')
    plt.ylabel('annualised returns')
    plt.legend(labelspacing=0.8)
```

### Frontera Eficiente

Se trata del conjunto de carteras más eficiente y rentable que podemos ver que se comercializan debido a que la rentabilidad esperada es mayor con respecto a otras carteras que no lo son. Esto se mide viendo y comparando los diferentes niveles de riesgo que asumen los particulares por invertir en determinadas carteras.

¿Por qué es eficiente?
- Debido a que los puntos a lo largo de la línea darán el menor riesgo para un retorno objetivo determinado.
- Todos los otros puntos directamente a la línea darán un mayor riesgo con los mismos retornos.

Para ello se realizan los siguientes pasos:
- Simulando muchas posibles elecciones aleatorias.
- Se eligen los mejores (riesgo mínimo o rendimiento máximo ajustado al riesgo).

Esto también se puede implementar utilizando la función de optimización de Scipy. La función de optimización de Scipy está haciendo una tarea similar para resolver en Excel cuando se le da qué optimizar y cuáles son las restricciones y los límites.

Las siguientes funciones son para obtener la cartera máxima de relación de Sharpe:

- En la función de optimización de Scipy, no hay "maximizar", por lo que como una función objetiva debe pasar algo que debe minimizarse.

- Es por eso que el primer "neg_sharpe_ratio" es calcular la relación negativa de Sharpe. Ahora podemos usar esto como nuestra función objetivo para minimizar.

- En la función "max_sharpe_ratio", primero define argumentos (esto no debe incluir las variables que desea cambiar para la optimización, en este caso, "pesos").

"Límites" está dando otro límite para asignar ponderaciones aleatorias, al decir que cualquier ponderación debe estar comprendida entre 0 y 1. No se puede asignar una asignación presupuestaria negativa a una acción o una asignación superior al 100% a una acción.

```python
def neg_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate):
    p_var, p_ret = portfolio_annualised_performance(weights, mean_returns, cov_matrix)
    return -(p_ret - risk_free_rate) / p_var

def max_sharpe_ratio(mean_returns, cov_matrix, risk_free_rate):
    num_assets = len(mean_returns)
    args = (mean_returns, cov_matrix, risk_free_rate)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bound = (0.0,1.0)
    bounds = tuple(bound for asset in range(num_assets))
    result = sco.minimize(neg_sharpe_ratio, num_assets*[1./num_assets,], args=args,
                        method='SLSQP', bounds=bounds, constraints=constraints)
    return result
```

```python
def portfolio_volatility(weights, mean_returns, cov_matrix):
    return portfolio_annualised_performance(weights, mean_returns, cov_matrix)[0]

def min_variance(mean_returns, cov_matrix):
    num_assets = len(mean_returns)
    args = (mean_returns, cov_matrix)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bound = (0.0,1.0)
    bounds = tuple(bound for asset in range(num_assets))

    result = sco.minimize(portfolio_volatility, num_assets*[1./num_assets,], args=args,
                        method='SLSQP', bounds=bounds, constraints=constraints)

    return result
```

Como ya mencionamos anteriormente, también podemos trazar una línea que describa dónde deberían estar las carteras eficientes para una tasa de riesgo dada. Esto se llama "frontera eficiente". A continuación defino otras funciones para calcular la frontera eficiente. La primera función "Efficient_return "está calculando la cartera más eficiente para un rendimiento objetivo dado, y la segunda función "Efficient_frontier" tomará un rango de rendimientos objetivo y calculará la cartera eficiente para cada nivel de rendimiento.

```python
def efficient_return(mean_returns, cov_matrix, target):
    num_assets = len(mean_returns)
    args = (mean_returns, cov_matrix)

    def portfolio_return(weights):
        return portfolio_annualised_performance(weights, mean_returns, cov_matrix)[1]

    constraints = ({'type': 'eq', 'fun': lambda x: portfolio_return(x) - target},
                   {'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0,1) for asset in range(num_assets))
    result = sco.minimize(portfolio_volatility, num_assets*[1./num_assets,], args=args, method='SLSQP', bounds=bounds, constraints=constraints)
    return result


def efficient_frontier(mean_returns, cov_matrix, returns_range):
    efficients = []
    for ret in returns_range:
        efficients.append(efficient_return(mean_returns, cov_matrix, ret))
    return efficients
```

## Cartera con Paridad de Riesgo (Risk Parity)
El enfoque de paridad de riesgo para la construcción de cartera busca asignar el capital en una cartera basada en una base ponderada por riesgo. La asignación de activos es el proceso por el cual un inversor divide el capital en una cartera entre diferentes tipos de activos. La asignación de cartera tradicional es del 60% a acciones y del 40% a bonos. Sin embargo, esta asignación no funciona bien durante las caídas del mercado de valores y la inestabilidad económica. El enfoque de paridad de riesgo intenta evitar los riesgos y sesgos de la diversificación de cartera tradicional. Permite la construcción de una cartera óptima considerando la volatilidad de los activos incluidos en la cartera.

```python
from scipy.optimize import minimize

def target_risk_contributions(target_risk, cov):
    """
    Returns the weights of the portfolio that gives you the weights such
    that the contributions to portfolio risk are as close as possible to
    the target_risk, given the covariance matrix
    """
    n = cov.shape[0]
    init_guess = np.repeat(1/n, n)
    bounds = ((0.0, 1.0),) * n # an N-tuple of 2-tuples!
    # construct the constraints
    weights_sum_to_1 = {'type': 'eq',
                        'fun': lambda weights: np.sum(weights) - 1
    }
    def msd_risk(weights, target_risk, cov):
        """
        Returns the Mean Squared Difference in risk contributions
        between weights and target_risk
        """
        w_contribs = risk_contribution(weights, cov)
        return ((w_contribs-target_risk)**2).sum()
    
    weights = minimize(msd_risk, init_guess,
                       args=(target_risk, cov), method='Powell',
                       options={'disp': False},
                       constraints=(weights_sum_to_1,),
                       bounds=bounds)
    return weights.x


def equal_risk_contributions(cov):
    """
    Returns the weights of the portfolio that equalizes the contributions
    of the constituents based on the given covariance matrix
    """
    n = cov.shape[0]
    return target_risk_contributions(target_risk=np.repeat(1/n,n), cov=cov)

```
